```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Q1. import and implement load_breast_cancer() dataset.
data_cancer = load_breast_cancer()
#print(data_cancer.DESCR)

# Q2. Explain load_breast_cancer dataset, and tell what are the classes, features and samples of this dataset.
# Features and labels
X = data_cancer.data  # this is the features
y = data_cancer.target  # this is the target

df = pd.DataFrame(X, columns=data_cancer.feature_names)
df['target'] = y

# Print out the first few samples
print("sample from the breast cancer dataset:\n")
for i in range(1):
    print(f"Sample {i + 1}:")
    print(f"Features: {df.iloc[i, :-1].values}")
    print(f"Target: {'Malignant' if df.iloc[i, -1] == 0 else 'Benign'}\n")

# Split the dataset into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Q3. Train and test the GaussianNB algorithm
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)
gnb_accuracy = accuracy_score(y_test, y_pred_gnb)
print(f"GaussianNB Accuracy: {gnb_accuracy:.4f}")
```

```python
# Q4. Train K Nearest Neighbors algorithm on 70% of the data and test it on 30% of the data set
# Function to train KNN for different k values
def train_knn(k):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_knn = knn.predict(X_test)
    return accuracy_score(y_test, y_pred_knn)
# a, Run the algorithm for k values, k = 3, k= 5, k = 7 and k = 9
k_values = [3, 5, 7, 9]
accuracies_knn = {}

# Evaluate KNN for each k value
for k in k_values:
    accuracy_knn = train_knn(k)
    accuracies_knn[k] = accuracy_knn
    print(f"KNN Accuracy for k={k}: {accuracy_knn:.4f}")
```

b, What k returns better accuracy?

when k = 9, it returns better accuracy.

```python
# Q5. Show the accuracy of the two algorithms
# Q6. Which algorithm has better accuracy?
best_k = max(k_values, key=lambda k: train_knn(k))
best_knn_accuracy = train_knn(best_k)

# Print the best k value and its accuracy
print(f"GaussianNB Accuracy: {gnb_accuracy:.4f}")
print(f"\nBest K for KNN: {best_k} with Accuracy: {best_knn_accuracy:.4f}")

# Compare accuracies and determine which algorithm has better accuracy
if gnb_accuracy > best_knn_accuracy:
    print("\nGaussianNB has better accuracy.")
else:
    print("\nK-Nearest Neighbors has better accuracy.")
```
