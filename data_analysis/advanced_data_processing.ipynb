```python
# Necessary libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

# 1. Load the breast-cancer dataset
data_cancer = load_breast_cancer()

# 2. Create the pandas df from the above dataset
df = pd.DataFrame(data=data_cancer.data, columns=data_cancer.feature_names)
df['target'] = data_cancer.target

# 3. Display the dataset into a pandas DataFrame
# Display the DataFrame's head
print("\n\n-----------------------------DataFrame Head-------------------------\n\n")
print(df.head())

# Display the DataFrame's tail
print("\n\n\n---------------------------DataFrame Tail---------------------------\n")
print(df.tail())

# 4. Use the DataFrame method describe to calculate the descriptive statistics for the numerical data columns
print("\nDescriptive Data Statistics:")
print(df.describe())

# 5. Use matplotlib to view histograms of each numerical data column.
df.drop('target', axis=1).hist(figsize=(10, 14), bins=10)
plt.suptitle('\nHistograms of Numerical ')
plt.show()
```
