```python
import numpy as np

# Create an array with values 1 to 15
arr = np.arange(1, 16)

# Reshape it into a 3-by-5 array
arr = arr.reshape(3, 5)

#a) Select row 2
row_2 = arr[1]

#b) Select the column 4.
column_4 = arr[:, 3]

#c) Select rows 0 and 1.
rows_0_and_1 = arr[:2]

#d) Select columns 2–4.
columns_2_to_4 = arr[:, 1:5]

#e) Select the element that is in row1 and column4
element_row1_column4 = arr[1, 4]

#f) Select all elements from rows 1 and 2 that are in columns 0, 2 and 4.
rows_1_and_2_columns_0_2_4 = arr[1:3, [0, 2, 4]]
```

```python
# Printing out the results

print("a) Select row 2 \n\n Row 2:", row_2)

print("\n b) Select the column 4\n Column 4:", column_4)

print("\n c) Select rows 0 and 1\n Rows 0 and 1:\n", rows_0_and_1)

print("\n d) Select columns 2–4\n Columns 2 to 4:\n", columns_2_to_4)

print("\n e) Select the element that is in row 1 and column 4\n Element in row1 and column4:", element_row1_column4)

print("\n f) Select all elements from rows 1 and 2 that are in columns 0, 2 and 4\n Selected elements:\n", rows_1_and_2_columns_0_2_4)
```

```python
import numpy as np

# Define the arrays
array1 = np.array([[0, 1], [2, 3]])
array2 = np.array([[4, 5], [6, 7]])

# a) Use vertical stacking to create the 4-by-2 array named array3 with array1 stacked on top of array2
array3 = np.vstack((array1, array2))

# b) Use horizontal stacking to create the 2-by-4 array named array4 with array2 to the right of array1
array4 = np.hstack((array1, array2))

# c) Use vertical stacking with two copies of array4 to create a 4-by-4 array named array5
array5 = np.vstack((array4, array4))

# d) Use horizontal stacking with two copies of array3 to create a 4-by-4 array named array6
array6 = np.hstack((array3, array3))

```

```python
# Printing out the results
print("\n a) Vertical stacking of array1 on top of array2\n Array3:\n", array3)

print("\n b) Horizontal stacking of array1 and array2\n Array4:\n", array4)

print("\n c) Vertical stacking of two copies of array4\n Array5:\n", array5)

print("\n d) Horizontal stacking of two copies of array3\n Array6:\n", array6)
```

```python
import numpy as np

# Createing heckerboard pattern
pattern = np.array([['_', '*'], ['_', '*']])

checkerboard = np.tile(pattern, (6, 8))

# Print the checkerboard pattern
print('\n'.join([' '.join(row) for row in checkerboard]))
```
