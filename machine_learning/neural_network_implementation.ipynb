Part 1:    
Consider a neuron with 2 inputs, 1 output, and a threshold activation function. If the two weights are
w1=−1 and w2=−1, and the bias is b=1.5, then what is the output for input (0 , 0)?    
 What about for
inputs (1 , 0), (0 , 1), and (1 , 1)?       


Answer part 1:  

Weights:- w1 = -1, w2 = =1
Bias:- b = 1.5

If z > or = 0, the output is = 1, otherwise the output is = 0.


1. input (0 ,0)  
     z = -1 * 0 - 1 * 0 + 1.5 = 1.5
      Output = 1

2. input (1 ,0)  
     z = -1 * 1 - 1 * 0 + 1.5 = 0.5
      Output = 1
  
3. input (0 ,1)  
     z = -1 * 0 - 1 * 1 + 1.5 = 0.5
      Output = 1

1. input (1 ,1)  
     z = -1 * 1 - 1 * 1 + 1.5 = -0.5
      Output = 0




```python
# Q1 part 2: Draw the discriminant function for this function and write down its equation.

import numpy as np
import matplotlib.pyplot as plt

# -----------------------
# Section 1: Threshold Activation Perceptron for NAND Logic
# -----------------------

# Define weights and bias for the perceptron
w1, w2 = -1, -1
b = 1.5

# Define the threshold activation function
def threshold_activation(x1, x2):
    z = w1 * x1 + w2 * x2 + b
    return 1 if z >= 0 else 0

# Define the inputs for the NAND function
inputs = [(0, 0), (1, 0), (0, 1), (1, 1)]

# Calculate outputs using the threshold activation function
outputs = [(x1, x2, threshold_activation(x1, x2)) for x1, x2 in inputs]

# Display results from the perceptron
print("Threshold Activation (Perceptron) Outputs:")
for x1, x2, output in outputs:
    print(f"Input: ({x1}, {x2}) -> Output: {output}")

# -----------------------
# Plot the Discriminant Function for Threshold Activation (Perceptron)
# -----------------------

# Discriminant function line: w1*x1 + w2*x2 + b = 0
x = np.linspace(-0.5, 2, 100)
y_discriminant = 1.5 - x  # This corresponds to x1 + x2 = 1.5 (discriminant function)

# Plot the discriminant line
plt.figure(figsize=(6, 6))
plt.plot(x, y_discriminant, label="Discriminant Function: x1 + x2 = 1.5", color='blue')

# Plot the input points based on the perceptron outputs
colors = ['red', 'green', 'green', 'green']  # Color inputs based on threshold output
plt.scatter([0, 1, 0, 1], [0, 0, 1, 1], c=colors, label="Inputs")

# Set plot limits, labels, and display grid
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Discriminant Function and Input Points (Perceptron)")
plt.show()
```

Write a program that constructs logical NAND using neural networks, you can use the predefined modules in scikit learn.

Answer Part 2:   

```python
import numpy as np
from sklearn.neural_network import MLPClassifier

# -----------------------
# Section 2: Neural Network Training (MLP) for NAND Logic
# -----------------------

# Define the input for the NAND function
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Define the output for the NAND function
y = np.array([1, 1, 1, 0])

# Initialize and train a Multi-layer Perceptron (MLP) classifier
clf = MLPClassifier(hidden_layer_sizes=(2,), activation='relu', max_iter=1000, random_state=42)
clf.fit(X, y)

# Get the predictions on the input data
predictions = clf.predict(X)

# Display the input-output table comparing NAND outputs and neural network predictions
print("\nNeural Network (NAND) Output vs Predicted Output:")
print("Input (x1, x2) | NAND Output | Predicted Output")
print("-" * 40)
for i in range(len(X)):
    print(f"   {tuple(X[i])}     |      {y[i]}      |       {predictions[i]}")
print("-" * 40)

# -----------------------
# Display the trained weights and bias of the neural network
# -----------------------

# Get the trained weights and bias from the neural network model
weights = clf.coefs_
bias = clf.intercepts_

# Display the trained weights and biases
print("\nTrained Weights:")
for i, weight in enumerate(weights):
    print(f" Layer {i+1} weights: {weight}")

print("\nTrained Bias:")
for i, b in enumerate(bias):
    print(f" Layer {i+1} bias: {b}")
```

Suppose we have a system that observes a person’s TV watching habits to recommend other TV shows
the person may like. Suppose that we have characterized each show by whether it is a comedy, doctors,

    1. What are the features, and targets in this training dataset? How many classes does this example have?   

    2. Use Support Vector Machines on the training dataset, to predict whether the user will like or not like the TV show with the attributes Comedy = true, Doctors = true, Lawyers = False, Guns =  False

Answers Q3 part 1:  


1. What are the features, and targets in this training dataset? How many classes does this example have?  

The features in the dataset are:

Doctors    
Lawyers  
Guns   

The Targets are the LIKES    

How many classes does this example have?  

There are two classes:-   

   Class 1 (Likes = True): The person likes the show.   
   Class 2 (Likes = False): The person does not like the show.   





```python
# Q3 part 2:  Use Support Vector Machines on the training dataset, to predict whether the user will like or not like the TV show with the attributes Comedy = true, Doctors = true, Lawyers = False, Guns =  False
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Create the dataset based on the given table
X = np.array([
    [False, True, False, False],
    [True, False, True, False],
    [False, False, True, True],
    [False, False, True, False],
    [False, False, False, True],
    [True, False, False, True],
    [True, False, False, False],
    [False, True, True, True],
    [False, True, True, False],
    [True, True, True, False],
    [True, True, False, True],
    [False, False, False, False]
])
#Target (y): The Likes column,
y = np.array([False, True, True, False, False, False, True, True, False, True, False, False])

# Split data into training and test sets (though we have a small dataset, we'll train on all data)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the SVM model
svm_model = SVC(kernel='linear')  # Linear kernel
svm_model.fit(X_train, y_train)

# Prediction on the new input: Comedy=True, Doctors=True, Lawyers=False, Guns=False
new_data = np.array([[True, True, False, False]])  # The input vector

# Predict whether the person likes the TV show
prediction = svm_model.predict(new_data)

# Output the prediction
#Q3 part 2: predict whether the user will like or not like the TV show with the attributes Comedy = true, Doctors = true, Lawyers = False, Guns = False
print(f"Prediction on the input: Comedy=True, Doctors=True, Lawyers=False, Guns=False (Likes).\n\n    Prediction : {prediction[0]}")
```

Is Support Vector Machines a supervised or unsupervised learning? Explain why.
   
Support Vector Machines (SVM) is a supervised learning algorithm because it uses labeled data during training. The model learns a decision boundary (or hyperplane) to classify data into predefined categories based on the provided input-output pairs.
