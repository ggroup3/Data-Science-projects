Q 2.	Explain what is the curse of dimensionality? What is the relation between machine learning problems and the curse of dimensionality?

Answer:    
	The curse of dimensionality means that as the number of features (dimensions) in data increases, it becomes harder for machine learning models to find patterns.   
	This can lead to problems like overfitting, computational inefficiency, and decreased accuracy in machine learning models.



Q 3.	Explain the process of supervised learning.

Answer:    
	Supervised learning is a process where a model is trained on labeled data. We collect data that has input-output pairs (labels) than the model learns by mapping inputs to the correct outputs using this labeled data and model’s accuracy is measured by comparing its predictions to the actual labels.


Q 4. Write a python function named numVowels that is passed a string containing letters, each of which may be in either uppercase or lowercase, and returns a tuple containing the number of vowels and the number of constants the string contains.


```python
# Answer Q4
def numVowels(s):
    vowels = "aeiou"
    num_vowels = 0
    num_consonants = 0

    for letter in s:
        if letter.isalpha():
          # Count vowels
            if letter.lower() in vowels:
                num_vowels += 1
            else:
                num_consonants += 1

    return num_vowels, num_consonants

letters = "Word"
Vowels_And_Consonants = numVowels(letters)
print(f"{letters}\nVowels: {Vowels_And_Consonants[0]}\nConsonants: {Vowels_And_Consonants[1]}")
```

Q 5. Perform the following tasks with pandas Series:
Create a Series from the list [7,11,13,17]
Create a Series with five elements that are all 100.0
Create a Series with 20 elements that are all random numbers in the range 0 to 100. Use the method describe to produce the Series’ basic descriptive statistics.
Create a Series called grades of the floating-point values 98.6, 77.9, 100, and 83.4 using the index keyword specify the custom indices ‘Julie’, ’Charlie’, ‘Sam’, and ‘Andera’

```python
# Answer Q5

import pandas as pd
import numpy as np

# a. Create a Series from the list [7, 11, 13, 17]
A_series = pd.Series([7, 11, 13, 17])
print("a. Series from list [7, 11, 13, 17]:")
print(A_series)

# b. Create a Series with five elements that are all 100.0
B_series = pd.Series([100.0] * 5)
print("\nb. Series with five elements that are all 100.0:")
print(B_series)

# c. Create a Series with 20 random numbers in the range 0 to 100
C_series = pd.Series(np.random.randint(0, 101, size=20))
print("\nc. Series with 20 random numbers in the range 0 to 100:")
print(C_series)

# Series’ basic descriptive statistics
print("\nDescriptive statistics:")
print(C_series.describe())

# d. Create a Series called grades with custom indices
grades = pd.Series([98.6, 77.9, 100.0, 83.4], index=['Julie', 'Charlie', 'Sam', 'Andera'])
print("\nd. grades of the floating-point values 98.6, 77.9, 100, and 83.4:")
print(grades)
```

Q 6. Write the following functions:           
a.  def firstDigit(n) returning the first digit of the argument      
b.  def lastDigit(n) returning the last digit of the argument    
c.  def digits(n) returning the numbers of digits in the argument  


```python
# Answer Q6

# a. def firstDigit(n) returning the first digit of the argumenta
def firstDigit(n):
    return int(str(n)[0])
# b. def lastDigit(n) returning the last digit of the argument
def lastDigit(n):
    return int(str(n)[-1])
# c. def digits(n) returning the numbers of digits in the argument
def digits(n):
    return len(str(n))


n = 24810
print("a. First digit:", firstDigit(n))
print("b. Last digit:", lastDigit(n))
print("c. Number of digits:", digits(n))
```
